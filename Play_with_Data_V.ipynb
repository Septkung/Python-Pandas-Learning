groupby คือการจัดการกับกลุ่มของข้อมูล

```python
df.groupby('ชื่อเทศกาล').count()['รหัสจังหวัด'] #output is a DataFrame
#can give command more specific Ex. ['รหัสจังหวัด']
```

```python
df.groupby('จังหวัด').count()['ชื่อเทศกาล'].sort_values(ascending=False)[0:20]
# sort_values() has ascending=True as a default
# And you use slice command to see some row Ex.[0:20] for 0-20
```

groupby Unit 3

```python
#use Covid-19 Data
df_II = pd.read_csv('thailand_data.csv')
```

```python
df_II.columns
```

```python
df_II['announce_date'] = pd.to_datetime(df_II['announce_date'],dayfirst=True)
```

```python
df_II.dtypes
```

```python
import datetime

filtered_df = df_II[df_II['announce_date'].dt.date >= datetime.date(year=2021,month=4,day=1)]
```

```python
filtered_df.head()
```

```python
filtered_df.groupby('announce_date').count()['No.']
```

```python
filtered_df.groupby(['province_of_isolation','announce_date']).count()['No.']
#Index is 'province_of_isolation','announce_date' and Count for No.
```

groupby reset_index and pivot

```python
count_df = filtered_df.groupby(['province_of_isolation','announce_date']).count()['No.'].reset_index()
# reset_index will reset index let we work more easier
```

```python
count_df
```

Pivot Command use for reformat DataFrame

```python
count_df = count_df.pivot(index='province_of_isolation',columns='announce_date',values='No.').fillna(0)
# we want to reformat DataFrame 
# เราจึงจะให้ index เป็น จังหวัด
# ให้ columns เป็นวันที่ ประกาศ
# ให้ values เป็น จำนวนนับของผู้ติดเชื้อ
```

```python
count_df
```
