Data Vitualization

```python
import pandas as pd 

df = pd.read_csv("thailand_data.csv")
```

```python
df['announce_date'] = pd.to_datetime(df['announce_date'],dayfirst=True)
```

```python
import datetime

df = df[df['announce_date'].dt.date >= datetime.date(year=2021,month=4,day=1)]
```

```python
df
```

```python
from matplotlib import font_manager as fm
import matplotlib.pyplot as plt

#load thai font
fig,ax = plt.subplots()
fontprop = fm.FontProperties(fname="THSarabunNew.ttf")
show_df = df['province_of_isolation'].value_counts()
ax.set_xticklabels(show_df.index,fontproperties=fontprop,fontsize=12)

show_df.plot(kind='bar',figsize=(10,6))
```

```python
df['province_of_isolation'].value_counts().plot(kind='bar')
# In default cant use thai font 
```

```python
show_df = df['sex'].value_counts()

fig,ax = plt.subplots()
ax.set_xticklabels(show_df.index,fontproperties=fontprop)
show_df.plot(kind="bar",fontsize=24,color=['cyan','pink'])
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
filtered_df = df[df['age'] <= 120]
```

```python
count_df = filtered_df.groupby(['province_of_isolation','announce_date']).count()['No.'].reset_index().pivot(index='province_of_isolation',columns='announce_date',values='No.').fillna(0)
```

```python
count_df.head()
```

```python
count_df = count_df.T # use for tranpost row <--> columns 
```

```python
count_df['กรุงเทพมหานคร'].plot(kind='bar',figsize=(18,6))
```

```python
# รวมสะสม
count_df['กรุงเทพมหานคร'].cumsum().plot()
```

```python
ax = count_df.reset_index()['กรุงเทพมหานคร'].plot(kind='bar',figsize=(18,6))
# make bar chart 
ax = count_df.reset_index()['กรุงเทพมหานคร'].cumsum().plot(kind='line',ax=ax,color='red')
# make line chart from cumulate sumation data
t = ax.set_xticklabels([str(c).split()[0] for c in list(count_df.index)])
# cut the time out of announce_date
t = plt.xticks(rotation=90)
# rotate x axis 90 degree
```
