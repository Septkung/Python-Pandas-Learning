```python
import pandas as pd 

df = pd.read_csv('thailand_data.csv')
```

```python
import datetime

df['announce_date'] = pd.to_datetime(df['announce_date'],dayfirst=True)
filtered_df = df[df['announce_date'].dt.date >= datetime.date(year=2021,month=4,day=1)]
count_df = filtered_df.groupby(['province_of_isolation','announce_date']).count()['No.'].reset_index()
count_df = count_df.pivot(index='province_of_isolation',columns='announce_date',values='No.').fillna(0)
```

```python
count_df
```

```python
# ดูแนวโน้มการเพิ่ม/ลดของผู้ติดเชื้อ
count_df[list(count_df.columns)[-1]] # last record
count_df['2021-07-15'] # Or you can use this
```

```python
# We can minus between list and list
count_df[list(count_df.columns)[-1]] - count_df[list(count_df.columns)[-2]]
```

```python
count_df['diff'] = count_df[count_df.columns[-3]] - count_df[count_df.columns[-4]]
```

Normalization

```python
# Vector Processing Average
count_df['diff_percent'] = (count_df[count_df.columns[-3]] - count_df[count_df.columns[-4]])/count_df[count_df.columns[-4]]*100
```

```python
# รวมกรุงเทพและปริมณฑล
# declare function 
def convert_province(province):
    if province in ['กรุงเทพมหานคร','นนทบุรี','ปทุมธานี','สมุทรปราการ','นครปฐม','สมุทรสาคร']:
        return 'กรุงเทพและปริมณฑล'
    return province

df['province_of_isolation'].apply(convert_province)
```

```python
# Or you can use lamda function
df['province_of_isolation'].apply(lambda x: 'กรุงเทพและปริมณฑล' if x in ['กรุงเทพมหานคร','นนทบุรี','ปทุมธานี','สมุทรปราการ','นครปฐม','สมุทรสาคร'] else x)
```

```python
df['province_group'] = df['province_of_isolation'].apply(convert_province)
```
