```python
df.announce_date = pd.to_datetime(df.announce_date,dayfirst=True)
```

```python
df.dtypes
```

```python
# want row age == female
df[df.sex == 'หญิง']
# ถ้าเขียน df.sex=="หญิง" เฉยๆ มันจะไปดูที่ข้อมูลแล้วให้ค่า True\False แทน
```

```python
df.dtypes
```

```python
df.isna().sum()
# in pandas True = 1 ,False = 0
# we can find which columns have Nan value by this command 
```

```python
df.age.dropna()
# filter the data that have Nan in age columns
```

```python
df['filled_age'] = df.age.fillna(-1)
# Or fill Nan with -1 <imputation>
# should add new columns for avoiding forget 
```

```python
df.filled_age = pd.to_numeric(df.filled_age) #this command had a Error
```

```python
df.loc[100164] # check Columns that had Error
```

```python
df.filled_age = df.filled_age.replace("55/664",-1)
```

```python
df.loc[100164]
```

```python
df.filled_age = pd.to_numeric(df.filled_age).astype('int32')
```

```python
df.dtypes
```

```python
df[df.filled_age >= 60] #เลือกข้อมูลที่มีอายุมากกว่าเท่ากับ 60
```

```python
import datetime

#ต้องการแค่ข้อมูลหลังจาก 01/04
fillter_df = df[df['announce_date'].dt.date >= datetime.date(2021,4,1)]
#datetimeproperties object
#เวลาจะเรียกวันมาเปรียบเทียบต้องใช้ libary
```

```python
fillter_df
```

If we have mutiple condition

```python
result_df = fillter_df[(fillter_df['sex'] == 'ชาย') & (fillter_df['filled_age'] >= 1 ) & (fillter_df['filled_age'] < 99)]
# add condition in () and use operation symbol
# & คือ and
# | คือ or
```

```python
result_df
```

```python
result_df['filled_age'].value_counts()
#คำสั่งนับค่าที่อยู่ใน Series
```
